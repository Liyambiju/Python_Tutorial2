```python
import tkinter as tk
from tkinter import messagebox
import re

def check_strength():
    password = password_entry.get()

    if password == "":
        messagebox.showwarning("Input Error", "Password cannot be empty.")
        return

    length_ok = len(password) >= 8
    digit_ok = any(char.isdigit() for char in password)
    special_ok = bool(re.search(r"[!@#$%^&*(),.?\":{}|<>]", password))

    score = length_ok + digit_ok + special_ok

    if score == 1:
        result_var.set("Weak")
    elif score == 2:
        result_var.set("Moderate")
    elif score == 3:
        result_var.set("Strong")
    else:
        result_var.set("Weak")

# Window
root = tk.Tk()
root.title("Password Strength Checker")
root.geometry("350x200")

tk.Label(root, text="Enter Password:").pack(pady=5)
password_entry = tk.Entry(root, show="*")
password_entry.pack(pady=5)

tk.Button(root, text="Check Strength", command=check_strength).pack(pady=10)

tk.Label(root, text="Strength:").pack()
result_var = tk.StringVar()
result_entry = tk.Entry(root, textvariable=result_var, state="readonly")
result_entry.pack(pady=5)

root.mainloop()
```
