```python
import tkinter as tk
from tkinter import messagebox

def calculate(operation):
    try:
        num1 = int(entry1.get())
        num2 = int(entry2.get())

        if operation == "add":
            result = num1 + num2
        elif operation == "sub":
            result = num1 - num2
        elif operation == "mul":
            result = num1 * num2
        elif operation == "div":
            if num2 == 0:
                raise ZeroDivisionError
            result = num1 / num2

        result_var.set(result)

    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter valid integers.")
    except ZeroDivisionError:
        messagebox.showerror("Math Error", "Cannot divide by zero.")

# Window
root = tk.Tk()
root.title("Arithmetic Calculator")
root.geometry("350x250")

# Input Fields
tk.Label(root, text="Enter First Integer:").pack(pady=5)
entry1 = tk.Entry(root)
entry1.pack(pady=5)

tk.Label(root, text="Enter Second Integer:").pack(pady=5)
entry2 = tk.Entry(root)
entry2.pack(pady=5)

# Buttons Frame
button_frame = tk.Frame(root)
button_frame.pack(pady=10)

tk.Button(button_frame, text="Add", width=8,
          command=lambda: calculate("add")).grid(row=0, column=0, padx=5)

tk.Button(button_frame, text="Subtract", width=8,
          command=lambda: calculate("sub")).grid(row=0, column=1, padx=5)

tk.Button(button_frame, text="Multiply", width=8,
          command=lambda: calculate("mul")).grid(row=0, column=2, padx=5)

tk.Button(button_frame, text="Divide", width=8,
          command=lambda: calculate("div")).grid(row=0, column=3, padx=5)

# Output Field
tk.Label(root, text="Result:").pack(pady=5)
result_var = tk.StringVar()
result_entry = tk.Entry(root, textvariable=result_var, state="readonly")
result_entry.pack(pady=5)

root.mainloop()
```
