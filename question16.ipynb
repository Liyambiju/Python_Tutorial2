```python
import tkinter as tk
from tkinter import messagebox

def convert_temperature():
    try:
        celsius = float(celsius_entry.get())
        fahrenheit = (celsius * 9/5) + 32
        result_var.set(f"{fahrenheit:.2f}")  # Precision 2
    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter a valid numeric temperature.")

# Window
root = tk.Tk()
root.title("Temperature Converter")
root.geometry("350x200")

# Celsius Input
tk.Label(root, text="Enter Temperature (Celsius):").pack(pady=5)
celsius_entry = tk.Entry(root)
celsius_entry.pack(pady=5)

# Convert Button
tk.Button(root, text="Convert to Fahrenheit", command=convert_temperature).pack(pady=10)

# Result Field (Read-only)
tk.Label(root, text="Fahrenheit:").pack()
result_var = tk.StringVar()
result_entry = tk.Entry(root, textvariable=result_var, state="readonly")
result_entry.pack(pady=5)

root.mainloop()
```
