```python
import tkinter as tk
from tkinter import messagebox

def generate_bill():
    try:
        price = float(price_entry.get())
        quantity = int(quantity_entry.get())

        total = price * quantity

        if total > 1000:
            discount = total * 0.10
            final_amount = total - discount
        else:
            final_amount = total

        result_var.set(f"{final_amount:.2f}")

    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter valid numbers.")

# Window
root = tk.Tk()
root.title("Bill Generator")
root.geometry("350x250")

tk.Label(root, text="Item Price:").pack(pady=5)
price_entry = tk.Entry(root)
price_entry.pack(pady=5)

tk.Label(root, text="Quantity:").pack(pady=5)
quantity_entry = tk.Entry(root)
quantity_entry.pack(pady=5)

tk.Button(root, text="Generate Bill", command=generate_bill).pack(pady=10)

tk.Label(root, text="Final Amount:").pack()
result_var = tk.StringVar()
result_entry = tk.Entry(root, textvariable=result_var, state="readonly")
result_entry.pack(pady=5)

root.mainloop()
```
