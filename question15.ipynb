```python
import tkinter as tk

def clear_label():
    label.config(text="")
    clear_button.config(state="disabled")
    restore_button.config(state="normal")

def restore_label():
    label.config(text="Python GUI Demo")
    restore_button.config(state="disabled")
    clear_button.config(state="normal")

# Create window
root = tk.Tk()
root.title("GUI Demo")
root.geometry("300x150")

# Label
label = tk.Label(root, text="Python GUI Demo", font=("Arial", 14))
label.pack(pady=20)

# Buttons
clear_button = tk.Button(root, text="Clear", command=clear_label)
clear_button.pack(side="left", padx=30)

restore_button = tk.Button(root, text="Restore", command=restore_label, state="disabled")
restore_button.pack(side="right", padx=30)

root.mainloop()
```
